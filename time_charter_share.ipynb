```python
from random import randint

# Название порта, days_ballast, days_laden, days_load, days_disch, days_idle
burgas = {'name': 'Бургас', 
          'days_ballast': 1.55, 
          'days_laden': 1.55, 
          'days_load': 2.25, 
          'days_disch': 5, 
          'days_idle': 4}

odessa = {'name': 'Одесса', 
          'days_ballast': 1.55, 
          'days_laden': 1.55, 
          'days_load': 2.25, 
          'days_disch': 2.25, 
          'days_idle': 2}

durres = {'name': 'Дюррес', 
          'days_ballast': 5, 
          'days_laden': 5, 
          'days_load': 2.25, 
          'days_disch': 2.25, 
          'days_idle': 2}

genova = {'name': 'Генуя',
          'days_ballast': 6,
          'days_laden': 6,
          'days_load': 2.25,
          'days_disch': 2.25,
          'days_idle': 2}

samsun = {'name': 'Yarimca', 
          'days_ballast': 1.55, 
          'days_laden': 1.55, 
          'days_load': 2.25, 
          'days_disch': 2.25, 
          'days_idle': 2}

teribolu = {'name': 'Териболу', 
          'days_ballast': 1.55, 
          'days_laden': 1.55, 
          'days_load': 2.25, 
          'days_disch': 2.25, 
          'days_idle': 2}

mangalia = {'name': 'Мангалия',
            'days_ballast': 1.55,
            'days_laden': 1.55,
            'days_load': 2.25,
            'days_disch': 2.25,
            'days_idle': 2}

midia = {'name': 'Мидия',
         'days_ballast': 1.55,
         'days_laden': 1.55,
         'days_load': 2.25,
         'days_disch': 2.25,
         'days_idle': 2}

bur_mid = {'name': 'Бургас + Мидия',
           'days_ballast': 1.55,
           'days_laden': 2,
           'days_load': 2.25,
           'days_disch': 6,
           'days_idle': 2}

mid_man = {'name': 'Мидия + Мангалия',
           'days_ballast': 1.55,
           'days_laden': 2,
           'days_load': 2.25,
           'days_disch': 4,
           'days_idle': 2}


sam_ter = {'name': 'Самсун + Териболу', 
          'days_ballast': 1.55, 
          'days_laden': 2, 
          'days_load': 2.25, 
          'days_disch': 4.5, 
          'days_idle': 4}

# Вводные данные
daily_hire = 11500
monthly_hire = 345000
intake = 3500
IFO_price = 450
MGO_price = 600
euro = 1.22

# Сумма дисбурсментов
temryuk_da = 11773
durres_da = 6738 * euro
burgas_da = 11053 * euro
midia_da = 10796 * euro
mangalia_da = 10212 * euro
samsun_da = 8857
tirebolu_da = 16804
odessa_da = 30000
turkish_straits = 3132

# Потребление мазута
IFO_ballast = 11.5
IFO_laden = 11.5
IFO_loading = 1.5
IFO_disch = 1.5
IFO_idle = 1.8

# Потребление дизеля
MGO_ballast = 2
MGO_laden = 2
MGO_loading = 1.5
MGO_disch = 1.5
MGO_idle = 0.1



def calculate_freight(route):
    
    # Счетчики основных показателей (стоимости бункера, количества затраченных дней, суммы дисбура) 
    total_bunker_price = 0
    total_days = 0
    total_da = 0
    
    result = {}
   
    # Расчет стоимости фрахта за весь маршрут в течение года
    for i in range(len(route)):
        
        # Расчет потребления бункера
        IFO_consumption = route[i]['days_ballast']*IFO_ballast + route[i]['days_laden']*IFO_laden + route[i]['days_load']*IFO_loading + route[i]['days_disch']*IFO_disch + route[i]['days_idle']*IFO_idle 
        MGO_consumption = route[i]['days_ballast']*MGO_ballast + route[i]['days_laden']*MGO_laden + route[i]['days_load']*MGO_loading + route[i]['days_disch']*MGO_disch + route[i]['days_idle']*MGO_idle

        IFO_total_price_one_vyg = IFO_price * IFO_consumption 
        MGO_total_price_one_vyg = MGO_price * MGO_consumption

        # Расчет стоимости бункера
        total_bunker_price += IFO_total_price_one_vyg + MGO_total_price_one_vyg
        
        # Вывод результата за i-th маршрут 
        route_name = route[i]['name']
        print(f'Маршрут Темрюк / {route_name} / Темрюк')
        print(f'Кумулятивная сумма бункера = ' + str(total_bunker_price) + ' USD')
        total_days += route[i]['days_ballast'] + route[i]['days_laden'] + route[i]['days_load'] + route[i]['days_disch'] + route[i]['days_idle']
        print(f'Прошло дней: {total_days}')
        print()
        
        # Доп расходы в Бургасе
        if 'Бургас' in route[i]['name']:
            total_da += temryuk_da + burgas_da
        
        # Доп расходы в Одессе
        elif 'Одесса' in route[i]['name']:
            total_da += temryuk_da + odessa_da
        
        # Доп расходы в Дюрессе
        elif 'Дюррес' in route[i]['name']:
            total_da += temryuk_da + durres_da + turkish_straits
        
        # Доп расходы в Samsune               
        elif 'Самсун' in route[i]['name']:
            total_da += temryuk_da + samsun_da
        
        # Доп расходы в Териболу
        elif 'Териболу' in route[i]['name']:
            total_da += temryuk_da + tirebolu_da
        
        # Доп расходы в 2 портах выгрузки (Самсун + Териболу)
        elif 'Самсун + Териболу' in route[i]['name']:
            total_da += temryuk_da + samsun_da + tirebolu_da
            
        # Доп расходы в 2 портах выгрузки (Бургас + Мидия)
        elif 'Бургас + Мидия' in route[i]['name']:
            total_da += temryuk_da + burgas_da + midia_da
            
        # Доп расходы в 2 портах выгрузки (Мидия + Мангалия)
        elif 'Мидия + Мангалия' in route[i]['name']:
            total_da += temryuk_da + midia_da + mangalia_da
    
    
    total_unused_days = 243-total_days
    total_bunker_price += total_unused_days * (IFO_price * IFO_idle + MGO_price * MGO_idle)
    total_price = daily_hire*total_days + total_bunker_price + total_da
    total_price_per_mt = total_price / (intake*len(route))
    total_unused_freight = -1*(total_unused_days*daily_hire)
    print('---------------------------------------------------------------------------------------')
    
    if len(route) == 16:
        # Если 2 вояджа в месяц
        total_gross_price = monthly_hire * 8 + total_bunker_price + total_da 
        total_gross_price_per_mt = total_gross_price / (intake*len(route))
        print('БЕЗ УЧЕТА "БОНУСНЫХ" ДНЕЙ')
        print('Расчетная стоимость фрахта за весь период:           2760000.00 USD')
        print('Расчетная стоимость бункера (200000*мес):            1600000.00 USD')
        print('Расчетная стоимость дисбурсментов и др.:             215755.00 USD')
        print('Расчетная сумма валового фрахта:                     4575754.00 USD')
        print('Расчетная сумма валового фрахта на тонну:            77.29 USD / т')        
        print()
        
    elif len(route) == 20:
        print('Расчетная стоимость фрахта за весь период:   2760000.00 USD')
        print('Расчетная стоимость бункера (200000*мес):    1600000.00 USD')
        print('Расчетная стоимость дисбурсментов и др.:     215755.00 USD')
        print('Расчетная стоимость проекта:                 4575754.00 USD')    
    
    print('С УЧЕТОМ "БОНУСНЫХ" ДНЕЙ')
    print(f'Стоимость валового фрахта за весь период составила:  {total_gross_price} USD')
    print(f'Стоимость валового фрахта на тонну за весь период:   {total_gross_price_per_mt} USD')
    print(f'Стоимость бункера за весь период составила:          {total_bunker_price} USD')
    print(f'Стоимость дисбурсментов и доп расходов:              {total_da} USD')
    print(f'Стоимость чистого фрахта составила:                  {total_price} USD')
    print(f'Стоимость чистого фрахта на тонну:                   {total_price_per_mt} USD / т')
    print(f'Осталось неиспользованных дней:                      {total_unused_days} дней')
    print(f'Неиспользованная сумма за год:                       {total_unused_freight} USD')
    result['total_bunker_price'] = total_bunker_price
    result['total_gross_price'] = total_gross_price
    result['total_price'] = total_price
    result['total_price_per_mt'] = total_price_per_mt
    result['total_unused_days'] = total_unused_days
    result['total_unused_freight'] = total_unused_freight
    return result
       

def calculate_freight_risk(route):
    
    # Счетчики основных показателей (стоимости бункера, количества затраченных дней, суммы дисбура) 
    total_bunker_price = 0
    total_idle_days = 0
    total_days = 0
    total_da = 0
    
    result = {}
    
    # Расчет стоимости
    for i in range(len(route)):
        IFO_consumption = route[i]['days_ballast']*IFO_ballast + route[i]['days_laden']*IFO_laden + route[i]['days_load']*IFO_loading + route[i]['days_disch']*IFO_disch + route[i]['days_idle']*IFO_idle 
        MGO_consumption = route[i]['days_ballast']*MGO_ballast + route[i]['days_laden']*MGO_laden + route[i]['days_load']*MGO_loading + route[i]['days_disch']*MGO_disch + route[i]['days_idle']*MGO_idle

        IFO_total_price_one_vyg = IFO_price * IFO_consumption 
        MGO_total_price_one_vyg = MGO_price * MGO_consumption

        total_bunker_price += IFO_total_price_one_vyg + MGO_total_price_one_vyg
        
        # Вывод результата за i-th маршрут 
        route_name = route[i]['name']
        
        # Случайное число дней простоя
        random_idle_days = randint(1,5)
        route[i]['days_idle'] = random_idle_days
        total_idle_days += random_idle_days
        
        print(f'Маршрут Темрюк / {route_name} / Темрюк с {random_idle_days} дней простоя')
        print(f'Кумулятивная сумма бункера = ' + str(total_bunker_price) + ' USD')
        total_days += route[i]['days_ballast'] + route[i]['days_laden'] + route[i]['days_load'] + route[i]['days_disch'] + route[i]['days_idle']
        print(f'Прошло дней: {total_days}')
        print()
        
        # Доп расходы в Бургасе
        if 'Бургас' in route[i]['name']:
            total_da += temryuk_da + burgas_da
        
        # Доп расходы в Одессе
        elif 'Одесса' in route[i]['name']:
            total_da += temryuk_da + odessa_da
        
        # Доп расходы в Дюрессе
        elif 'Дюррес' in route[i]['name']:
            total_da += temryuk_da + durres_da + turkish_straits
        
        # Доп расходы в Samsune               
        elif 'Самсун' in route[i]['name']:
            total_da += temryuk_da + samsun_da
        
        # Доп расходы в Териболу
        elif 'Териболу' in route[i]['name']:
            total_da += temryuk_da + tirebolu_da
        
        # Доп расходы в Мангалии
        elif 'Мангалия' in route[i]['name']:
            total_da += temryuk_da + mangalia_da
        
        # Доп расходы в 2 портах выгрузки (Самсун + Териболу)
        elif 'Самсун + Териболу' in route[i]['name']:
            total_da += temryuk_da + samsun_da + tirebolu_da
            
        # Доп расходы в 2 портах выгрузки (Мидия + Мангалия)
        elif 'Мидия + Мангалия' in route[i]['name']:
            total_da += temryuk_da + midia_da + mangalia_da
            
        # Доп расходы в Yarimca
        elif 'Yarimca' in route[i]['name']:
            total_da += temryuk_da + samsun_da + turkish_straits
            
        # Доп расходы в в Генуе
        elif 'Генуя' in route[i]['name']:
            total_da += temryuk_da + durres_da + turkish_straits
    
       
    total_unused_days = 243-total_days
    total_bunker_price += total_unused_days * (IFO_price * IFO_idle + MGO_price * MGO_idle)
    total_price = daily_hire*total_days + total_bunker_price + total_da
    total_price_per_mt = total_price / (intake*len(route))
    total_unused_freight = -1*(total_unused_days*daily_hire)
    print('---------------------------------------------------------------------------------------')
    
    if len(route) == 16:
        # Если 2 вояджа в месяц
        total_gross_price = monthly_hire * 8 + total_bunker_price + total_da 
        total_gross_price_per_mt = total_gross_price / (intake*len(route))
        print('БЕЗ УЧЕТА "БОНУСНЫХ" ДНЕЙ')
        print('Расчетная стоимость фрахта за весь период:           2760000.00 USD')
        print('Расчетная стоимость бункера (200000*мес):            1600000.00 USD')
        print('Расчетная стоимость дисбурсментов и др.:             215755.00 USD')
        print('Расчетная сумма валового фрахта:                     4575754.00 USD')
        print('Расчетная сумма валового фрахта на тонну:            77.29 USD / т')        
        print()
        
    elif len(route) == 20:
        print('Расчетная стоимость фрахта за весь период:   2760000.00 USD')
        print('Расчетная стоимость бункера (200000*мес):    1600000.00 USD')
        print('Расчетная стоимость дисбурсментов и др.:     215755.00 USD')
        print('Расчетная стоимость проекта:                 4575754.00 USD')    
    
    print('С УЧЕТОМ "БОНУСНЫХ" ДНЕЙ')
    print(f'Стоимость валового фрахта за весь период составила:  {total_gross_price} USD')
    print(f'Стоимость валового фрахта на тонну за весь период:   {total_gross_price_per_mt} USD')
    print(f'Стоимость бункера за весь период составила:          {total_bunker_price} USD')
    print(f'Стоимость дисбурсментов и доп расходов:              {total_da} USD')
    print(f'Стоимость чистого фрахта составила:                  {total_price} USD')
    print(f'Стоимость чистого фрахта на тонну:                   {total_price_per_mt} USD / т')
    print(f'Осталось неиспользованных дней:                      {total_unused_days} дней')
    print(f'Дополнительных дней простоя в среднем:               {total_idle_days / 16} дней')
    print(f'Неиспользованная сумма за год:                       {total_unused_freight} USD')
    result['total_bunker_price'] = total_bunker_price
    result['total_gross_price'] = total_gross_price
    result['total_price'] = total_price
    result['total_price_per_mt'] = total_price_per_mt
    result['total_unused_days'] = total_unused_days
    result['average_idle_days'] = total_idle_days / 16
    result['total_unused_freight'] = total_unused_freight
    return result
       
  
```
